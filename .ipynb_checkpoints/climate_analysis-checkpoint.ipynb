```python
#SQLAlchmey Dependencies
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy import Column, Integer, String, Float 
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func, extract

# Other Dependencies
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from dateutil.parser import parse 
```

```python
# Create engine
engine = create_engine("sqlite:///Resources/hawaii.sqlite", echo=False)
```

```python
# Reflect Database into ORM class
Base = automap_base()
Base.prepare(engine, reflect=True)
Measurements = Base.classes.measurements
Stations = Base.classes.stations
```

```python
# Use the Inspector to explore the database and print the table names
inspector = inspect(engine)
inspector.get_table_names()
columns = inspector.get_columns('stations')
for c in columns:
    print(c['name'], c["type"])
```

```python
Base.classes.keys()
```

```python
# Start a session to query the database
session = Session(engine)
```

```python
prcp_results = session.query(Measurements.date, Measurements.prcp).\
    filter(Measurements.date >= '2016-08-01', Measurements.date <= '2017-08-31').all()   
print(prcp_results)
```

```python
# Unpack date and prcp from results
date = [result[0] for result in prcp_results[:-1]]
prcp = [int(result[1]) for result in prcp_results[:-1]]
```

```python
# Create DataFrame and display stats
prcp_2017 = pd.DataFrame(prcp_results, columns=['date', 'precipitation'])
prcp_2017_group = prcp_2017.groupby(['date']).mean()
prcp_2017 = pd.DataFrame(prcp_2017_group).reset_index()
prcp_2017.set_index('date')
prcp_2017.describe()
```

```python
# Plot precip
prcp_2017.plot(x='date', y='precipitation', xticks=(1,50,100,150,200, 250, 300, 350), rot=45)
plt.title('Hawaiian Precepitation for Over 12 Months')
plt.savefig('Output/precip.png')
sns.set_style("darkgrid")
```

```python
# Query station count
station_count_results = session.query(Stations.station).count()
print(station_count_results)
```

```python
# Query measurements for station activity
station_activity = session.query(Measurements.station, func.count(Measurements.station)).\
    group_by(Measurements.station).order_by(func.count(Measurements.station).desc()).all()
print(station_activity)
sa_df = pd.DataFrame(station_activity, columns=['station', 'observation count'])
sa_df.head(9)
```

```python
# Query last 12 months of tob data for USC00519281
tobs_results = session.query(Measurements.tobs).\
    filter(Measurements.date >= '2016-08-01', Measurements.date <= '2017-08-31').\
    filter(Measurements.station == "USC00519281").all()
# print(tobs_results)
```

```python
# Unpack tobs_results into list
tobs = [result[0] for result in tobs_results[:-1]]
```

```python
# Create Histogram from tobs data
bins=12

plt.hist(tobs, bins=bins)
plt.ylabel('Frequency')
plt.xlabel('Temperature (F)')
plt.title("Temperature Frequency for 12 Month Period")
plt.savefig("Output/tobs_hist.png")
```

```python
# calc_temp function which returns min, max, avg temp for a given date range
# date formate YYYY-mm-dd
def calc_temp(start_date, end_date):
    calc_results = session.query(Measurements.date, Measurements.tobs).\
        filter(Measurements.date >= start_date, Measurements.date <= end_date).all()
        
    print("Query results:",calc_results)
        
#     Put results into DataFrame
    tobs = [result[1] for result in calc_results[:-1]]
    print("")
    print('Max:', np.max(tobs))
    print('Min:', np.min(tobs))
    print('Avg:', np.mean(tobs))
    
```

```python
# gather data using calc_temp
calc_temp("2017-06-01", "2017-06-03")
```

```python
# Create Bar Chart from calc_temp data
plt.bar(0,78.06, color='purple', yerr=7, align='edge', width=(.25))
plt.ylabel('Temp (F)')
plt.title('Avg Temp During Trip')
plt.xlim(-0.15, 0.45)
plt.xticks([])
plt.savefig("Output/temp_avg.png")
```

```python
# Query prcp totals for each station using data from 2017
station_prcp = session.query(Measurements.station, func.sum(Measurements.prcp)).\
    filter(func.strftime("%Y", Measurements.date) == "2017").\
    group_by(Measurements.station).order_by(func.sum(Measurements.prcp).desc()).all()
print(station_prcp)
```

```python
def daily_normals(date):
#     d = datetime.datetime.strptime(date, "%m-%d")
    d = parse(date)
    month = str(d.month)
    day = str(d.day)
    results = session.query(Measurements.tobs).\
    filter(extract('month', Measurements.date) == month).\
    filter(extract('day', Measurements.date)  == day).all()

    tobs = list(np.ravel(results))
    
    tobs_max = np.max(tobs)
    tobs_min = np.min(tobs)
    tobs_mean = np.mean(tobs)
    
    tobs_list = [f"{month}-{day}", tobs_max, tobs_min, tobs_mean]
    return tobs_list
    
```

```python
d = daily_normals("2017-06-01")
print(d)
```

```python
from collections import defaultdict
tobs_dic = defaultdict(list)
dates = ('06-02', "06-03", "06-04", "06-05", "06-06",)
for date in dates:
    day = daily_normals(date)
    tobs_dic['Date'].append(day[0])
    tobs_dic['Max Temp'].append(day[1])
    tobs_dic['Min Temp'].append(day[2])
    tobs_dic['Mean Temp'].append(day[3])
print(tobs_dic)
```

```python
tobs_df = pd.DataFrame(tobs_dic)
tobs_df.head()
```

```python
tobs_df.plot()
plt.title('Weather for Trip Window')
plt.xlabel('Dates')
plt.ylabel('Temp (F)')
plt.fill_between(tobs_df['Date'], tobs_df['Max Temp'], facecolor='blue', alpha=0.25)
plt.fill_between(tobs_df['Date'], tobs_df['Mean Temp'], facecolor='orange', alpha=0.25)
plt.fill_between(tobs_df['Date'], tobs_df['Min Temp'], facecolor='green', alpha=0.25)
plt.legend(frameon=True, loc=3, bbox_to_anchor=(0.02, 0.04))
plt.margins(x=0.02, y=0.04)
plt.savefig("Output/daily_normals.png")
```
